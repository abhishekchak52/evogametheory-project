```python
def gamma(i,N,w):
    Fi = (m*R*(i-1) + (S+(m-1)*P)*(N-i))/(N-1)
    Gi = ((T+(m-1)*P)*i + m*P*(N-i-1))/(N-1)
    fi = 1 - w + w*Fi
    gi = 1 - w + w*Gi
    return gi/fi
```

```python
R = 3
S = 0
T = 5
P = 1

m = 10 # number of rounds
w_list = [0.01, 0.1, 1]
population_sizes = np.concatenate((np.arange(10,500,20),np.arange(500,2000,100)))
for w in w_list:
    Nrho = []
    for N in population_sizes:
        rho = 1/(1+sum([product([gamma(i+1,N,w) for i in range(k+1) ]) for k in range(N-1)]))
        print(f'{N}\t{rho}\t{N*rho}')
        Nrho.append(N*rho)
    plot(population_sizes,Nrho,'.')
```
